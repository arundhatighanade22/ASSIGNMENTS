# Apriori Algorithm

```python
frequent_itemsets = apriori(movie, min_support=0.1, use_colnames=False)
frequent_itemsets
```

```python
data1= association_rules(frequent_itemsets, metric='lift' ,min_threshold=0.8)
data1
```

```python
data1[data1.lift>1]
```

```python
data1.sort_values('lift',ascending = False)[:]
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np

plt.scatter(data1['support'], data1['confidence'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
plt.scatter(data1['support'], data1['lift'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('lift')
plt.title('Support vs Lift')
plt.show()
```

```python
fit = np.polyfit(data1['lift'], data1['confidence'], 1)
fit_fn = np.poly1d(fit)
plt.plot(data1['lift'], data1['confidence'], 'yo', data1['lift'], 
 fit_fn(data1['lift']))
```
